```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
import random
import datetime


plt.style.use('fivethirtyeight')
```

```python
# Generate random data
np.random.seed(42)  

# Generate 20 random dates
start_date = datetime.date(2014, 1, 1)
end_date = datetime.date(2014, 12, 31)
date_data = [start_date + datetime.timedelta(days=random.randint(0, 364)) for _ in range(20)]

category_data = np.random.choice(['A', 'B', 'C'], 20)
numeric_data_1 = np.random.randint(1, 101, 20)
numeric_data_2 = np.random.randint(1, 101, 20)
float_data = np.random.uniform(0, 1, 20)

# Create DataFrame
df = pd.DataFrame({
    'date_column': date_data,
    'category_column': category_data,
    'numeric_column_1': numeric_data_1,
    'numeric_column_2': numeric_data_2,
    'float_column': float_data
})
```

```python
df.head(20)
```

```python
df["numeric_column_1"].hist(bins=10)
plt.title("Numeric Column 1 Distribution")
plt.grid(None)
plt.show()
```

```python
# Data Preparation
avg_df = df.groupby("category_column")["numeric_column_1"].mean()

# Bar Plot
avg_df.plot(
kind="bar",
rot=45,
title="Mean Numeric Column Value by Category Column"
)
plt.show()
```

```python
df.sort_values("date_column",inplace=True)

df.plot(x="date_column", 
y="numeric_column_1",
rot=45,
kind="line")
plt.show()
```

```python
df.plot(
    x="numeric_column_1",
    y="float_column",
    kind="scatter")
plt.title("Float Column vs. Numeric Column 1")
plt.show()
```

```python
df[df['category_column']=="A"]["numeric_column_1"].hist(alpha = 0.5)
df[df['category_column']=="B"]["numeric_column_1"].hist(alpha = 0.7)
plt.legend(["A","B"])
plt.show()
```
